```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# create the dataframe
df = pd.DataFrame()
df['column-1'] = np.linspace(1, 100, 5)
df['column-2'] = np.linspace(1, 100, 5)
df
```

```python
#scatter graph
#sns.lmplot( 'column-1','column-1' ,data=df)   # line +scatter
sns.lmplot('column-1','column-2',data=df,fit_reg=False)  # scatter graph
plt.show()
```

```python

x=[1,3,5,7]
y=[3,3,6,2]

sns.kdeplot(x,y)
plt.show()
```

```python
#bar graph
df['x'] = np.linspace(1, 100, 5)
df['y'] = np.linspace(1, 100, 5)
df.head()
sns.factorplot("x","y", data=df,kind="bar",palette="Greens",size=3,aspect=4)
plt.show()
df
```

```python
#bar graph
df['x'] = np.linspace(1, 100, 5)
df['y'] = np.linspace(1, 100, 5)
df.head()
sns.barplot("x","y", data=df,palette="Greens")
plt.show()
df
```
