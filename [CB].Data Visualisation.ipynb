```python
import matplotlib.pyplot as plt
import numpy as np
```

themes/styles are available in this package 

```python
themes=plt.style.available
print(themes)
```

```python
#Line Plot
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

Line Graph

```python
plt.plot(x,y1,color='red') #color is the option which will print a particular colour.
plt.plot(x,y2) #by default, it will print Blue colour 
plt.xlabel('X- axis') # X-label will print in the x axis
plt.ylabel('Y- axis') # Y-label will print in the y axis
plt.title('Line Graph') #title of the graph
plt.show() #prints the graph 
```

```python
plt.plot(x,y1,color='red',label="x,y1") #label is to print the sides in graph
plt.plot(x,y2,color='green',label='x,y2') #it accepts both "" and ''.
plt.xlabel('X- axis')  
plt.ylabel('Y- axis')
plt.title('Line Graph') 
plt.legend() #to print the labels in the graph.
plt.show() 
```

```python
plt.plot(x,'r+') #prints x line.R means red colou,r
plt.show()       #r+ means , + indicates in the x points in red color.
```

* indicates in the x points in default color(i.e., blue).

```python
plt.plot(x,'*') ## * indicates in the x points in default color(i.e., blue).
plt.show()
```

# Both x,r* and x in one graph

```python
plt.plot(x,'r*') 
plt.plot(x) 
plt.show()
```

#linestyle means the plotted line  will have  a different styles


```python
plt.plot(x,y1,color='red',label="x,y1",linestyle='dashed')
plt.plot(x,y2,color='green',label='x,y2')
plt.legend()
plt.show()
```

#
linewidth :- 
-> thickness/width of the line in the graph.
-> we can increase/decrease the width
-> by using this command,  (linewidth=5)

```python
plt.plot(x,y1,color='red',label="x,y1",linestyle='dashed',linewidth=3)
plt.plot(x,y2,color='green',label='x,y2',linewidth=5)
plt.legend()
plt.show()
```

Scatter Plots

```python
plt.scatter(x,y1,color='red',label="x,y1",marker='o')
plt.scatter(x,y2,color='green',label='x,y2',marker='*')
plt.legend()  #marker it pindicates a  symbol in the xpoints in the graph
plt.show()
```

Figure
** Adjust the Size of any plot(rows,columns).

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y1,color='red',label="x,y1",marker='o')
plt.scatter(x,y2,color='green',label='x,y2',marker='*')
plt.legend()  
plt.title("Scatter plot")
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

# width :- width of the line in gragh
                                                                                                                     #tick_label :- names of the particular data
                                                                                                                     [10,20,30] -> [gold,silver,platinum]

```python
xaxis=np.array([0,1,2])*3 
plt.bar(xaxis,[5,15,25],width=0.3,label='Current year',tick_label=['Silver','Platinum','Gold']) #Current year
plt.bar(xaxis+0.5,[10,20,30],width=0.5,label='Next year')#Next year
plt.legend()
plt.title("Bar Graph")
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

#
Background style in the graph using plt.style.use("#particular style") 

```python
plt.style.use("dark_background") #by default green&yelow color
xaxis=np.array([0,1,2])*3 
plt.bar(xaxis,[5,15,25],width=0.3,label='Current year',tick_label=['Silver','Platinum','Gold']) #Current year
plt.bar(xaxis+0.5,[10,20,30],width=0.5,label='Next year')#Next year
plt.legend()
plt.title("Bar Graph")
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

```python
plt.style.use("seaborn") 
xaxis=np.array([0,1,2])*3 
plt.bar(xaxis,[5,15,25],width=0.3,label='Current year',tick_label=['Silver','Platinum','Gold']) #Current year
plt.bar(xaxis+0.5,[10,20,30],width=0.5,label='Next year')#Next year
plt.legend()
plt.title("Bar Graph")
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

```python
plt.style.use("dark_background") 
xaxis=np.array([0,1,2])*3 
plt.bar(xaxis,[5,15,25],width=0.3,label='Current year',tick_label=['Silver','Platinum','Gold'],color='blue') #Current year
plt.bar(xaxis+0.5,[10,20,30],width=0.5,label='Next year',color='red')#Next year
plt.legend()
plt.title("Bar Graph")
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```
